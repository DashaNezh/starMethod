Пункт 1-2: Загрузка таблицы с сайта и оформление таблцы к "рабочему виду"


```python
# Загружаем данные с правильной ссылки на файл с таблицей
url = "https://pstu.ru/files/file/Abitur/2025%20final/%D0%A4%D0%9F%D0%9C%D0%9C_%D0%91_%D0%9D_000007052.html"

print(f"Загружаем данные с: {url}")

try:
    # Загружаем таблицы с помощью pd.read_html, пропускаем первые 11 строк
    tables = pd.read_html(url, skiprows=11)
    
    for i, table in enumerate(tables):
        print(f"\nТаблица {i+1}: {table.shape}")
        print(table.head())
        
except Exception as e:
    print(f"Ошибка при загрузке: {e}")
```

```python
# Обрабатываем загруженные данные
if 'tables' in locals() and len(tables) > 0:
    # Берем первую таблицу
    df = tables[0]
    
    # Устанавливаем строку 1 как заголовки
    df.columns = df.iloc[1]
    df = df.drop(df.index[0:2]).reset_index(drop=True)
    
    # Убираем лишние столбцы
    df = df.iloc[:, :-2]
    
    columns_to_remove = ['Идентификационный номер заказчика целевого обучения (для целевого приема)', 
                        'Номер предложения (для целевого приема)']
    df = df.drop(columns=columns_to_remove, errors='ignore')
    
    columns_to_remove = ['№', 'Уникальный код', 'Представление приказа', 'Учебная группа']
    df = df.drop(columns=columns_to_remove, errors='ignore')
    
    # Переименовываем индексы (начинаем с 1)
    df.index = range(1, len(df) + 1)
    
    pd.set_option('display.max_columns', None)
    pd.set_option('display.width', None)
    pd.set_option('display.max_colwidth', 15)
    pd.set_option('display.max_rows', None) #все строки
    
    display(df)
    
else:
    print("Не удалось загрузить таблицы")
```

Пункт 3: Проверка типов данных и приведение к числовому типу


```python
# Проверяем типы данных в столбцах
print("Типы данных в столбцах:")
print(df.dtypes)
print("\n" + "="*50)

# Показываем информацию о данных
print("\nИнформация о данных:")
print(df.info())
print("\n" + "="*50)

# Проверяем количественные столбцы и приводим к числовому типу
numeric_columns = ['Сумма баллов', 'АиНМА / Математика', 'ИТ / Информатика и ИКТ / Физика', 
                   'Русский язык', 'Сумма баллов за ИД']

print("\nПриводим количественные столбцы к числовому типу:")
for col in numeric_columns:
    if col in df.columns:
        print(f"Обрабатываем столбец: {col}")
        print(f"Тип до преобразования: {df[col].dtype}")
        
        # Преобразуем в числовой тип
        df[col] = pd.to_numeric(df[col], errors='coerce')
        
        print(f"Тип после преобразования: {df[col].dtype}")
        print(f"Количество пропущенных значений: {df[col].isna().sum()}")
        print("-" * 30)

print("\n" + "="*50)
print("Итоговые типы данных:")
print(df.dtypes)
```

Пункт 4: Статистика по суммарным баллам


```python
# Определяем статистику по суммарным баллам
print("Статистика по суммарным баллам:")
print("=" * 40)

# Средний балл
mean_score = df['Сумма баллов'].mean()
print(f"Средний суммарный балл: {mean_score:.2f}")

# Минимальный балл
min_score = df['Сумма баллов'].min()
print(f"Минимальный суммарный балл: {min_score}")

# Максимальный балл
max_score = df['Сумма баллов'].max()
print(f"Максимальный суммарный балл: {max_score}")
```

Пункт 5: Средние значения по предметам ЕГЭ


```python
# Определяем средние значения по предметам ЕГЭ
print("Средние значения по предметам ЕГЭ:")
print("=" * 50)

# Вычисляем средние значения по каждому предмету
math_mean = df['АиНМА / Математика'].mean()
it_mean = df['ИТ / Информатика и ИКТ / Физика'].mean()
russian_mean = df['Русский язык'].mean()

print(f"Средний балл по математике: {math_mean:.2f}")
print(f"Средний балл по ИТ/Информатике/Физике: {it_mean:.2f}")
print(f"Средний балл по русскому языку: {russian_mean:.2f}")

print("\n" + "=" * 50)

# Создаем новую таблицу для средних значений
subjects = ['Математика', 'ИТ/Информатика/Физика', 'Русский язык']
averages = [math_mean, it_mean, russian_mean]

# Создаем DataFrame с средними значениями
averages_df = pd.DataFrame({
    'Предмет': subjects,
    'Средний балл': averages
})

print("Новая таблица со средними значениями:")
display(averages_df)

print("\n" + "=" * 50)

# Определяем предмет с наибольшим средним баллом
max_subject = averages_df.loc[averages_df['Средний балл'].idxmax()]
print(f"Предмет с наибольшим средним баллом:")
print(f"Предмет: {max_subject['Предмет']}")
print(f"Средний балл: {max_subject['Средний балл']:.2f}")
```

Пункт 6: Определение абитуриентов типа "Тип_1"


```python
# Определяем абитуриентов типа "Тип_1"
print("Определение абитуриентов типа 'Тип_1':")
print("=" * 50)

# Вычисляем средние баллы по предметам
math_mean = df['АиНМА / Математика'].mean()
it_mean = df['ИТ / Информатика и ИКТ / Физика'].mean()
russian_mean = df['Русский язык'].mean()

print(f"Средний балл по математике: {math_mean:.2f}")
print(f"Средний балл по ИТ/Информатике/Физике: {it_mean:.2f}")
print(f"Средний балл по русскому языку: {russian_mean:.2f}")

print("\n" + "=" * 50)

# Определяем условия для типа "Тип_1"
# Русский язык ниже среднего, остальные предметы выше среднего
condition_type1 = (
    (df['Русский язык'] < russian_mean) &  # Русский ниже среднего
    (df['АиНМА / Математика'] > math_mean) &  # Математика выше среднего
    (df['ИТ / Информатика и ИКТ / Физика'] > it_mean)  # ИТ выше среднего
)

# Подсчитываем количество абитуриентов типа "Тип_1"
type1_count = condition_type1.sum()
print(f"Количество абитуриентов типа 'Тип_1': {type1_count}")

# Показываем абитуриентов типа "Тип_1"
type1_students = df[condition_type1]
print(f"\nАбитуриенты типа 'Тип_1':")
if len(type1_students) > 0:
    display(type1_students[['АиНМА / Математика', 'ИТ / Информатика и ИКТ / Физика', 'Русский язык', 'Сумма баллов']])
else:
    print("Абитуриентов типа 'Тип_1' не найдено")

print("\n" + "=" * 50)
print(f"Итого: {type1_count} абитуриентов из {len(df)} соответствуют типу 'Тип_1'")
print(f"Процент от общего числа: {(type1_count/len(df)*100):.1f}%")
```

Пункт 7: Определение абитуриентов типа "Тип_2"


```python
# Определяем абитуриентов типа "Тип_2"
print("Определение абитуриентов типа 'Тип_2':")
print("=" * 50)

# Вычисляем средние баллы по предметам
math_mean = df['АиНМА / Математика'].mean()
it_mean = df['ИТ / Информатика и ИКТ / Физика'].mean()
russian_mean = df['Русский язык'].mean()

print(f"Средний балл по математике: {math_mean:.2f}")
print(f"Средний балл по ИТ/Информатике/Физике: {it_mean:.2f}")
print(f"Средний балл по русскому языку: {russian_mean:.2f}")

print("\n" + "=" * 50)

# Определяем условия для типа "Тип_2"
# Русский язык выше среднего, остальные предметы ниже среднего
condition_type2 = (
    (df['Русский язык'] > russian_mean) &  # Русский выше среднего
    (df['АиНМА / Математика'] < math_mean) &  # Математика ниже среднего
    (df['ИТ / Информатика и ИКТ / Физика'] < it_mean)  # ИТ ниже среднего
)

# Подсчитываем количество абитуриентов типа "Тип_2"
type2_count = condition_type2.sum()
print(f"Количество абитуриентов типа 'Тип_2': {type2_count}")

# Показываем абитуриентов типа "Тип_2"
type2_students = df[condition_type2]
print(f"\nАбитуриенты типа 'Тип_2':")
if len(type2_students) > 0:
    display(type2_students[['АиНМА / Математика', 'ИТ / Информатика и ИКТ / Физика', 'Русский язык', 'Сумма баллов']])
else:
    print("Абитуриентов типа 'Тип_2' не найдено")

print("\n" + "=" * 50)
print(f"Итого: {type2_count} абитуриентов из {len(df)} соответствуют типу 'Тип_2'")
print(f"Процент от общего числа: {(type2_count/len(df)*100):.1f}%")
```

Пункт 8: Итоговая таблца


```python
# Вычисляем необходимые значения
total_students = len(df)
mean_total_score = df['Сумма баллов'].mean()
min_total_score = df['Сумма баллов'].min()
max_total_score = df['Сумма баллов'].max()

# Определяем предмет с высшим средним баллом
math_mean = df['АиНМА / Математика'].mean()
it_mean = df['ИТ / Информатика и ИКТ / Физика'].mean()
russian_mean = df['Русский язык'].mean()

if math_mean > it_mean and math_mean > russian_mean:
    best_subject = "Математика"
elif it_mean > russian_mean:
    best_subject = "ИТ/Информатика/Физика"
else:
    best_subject = "Русский язык"

# Подсчитываем количество студентов каждого типа
condition_type1 = (
    (df['Русский язык'] < russian_mean) &
    (df['АиНМА / Математика'] > math_mean) &
    (df['ИТ / Информатика и ИКТ / Физика'] > it_mean)
)

condition_type2 = (
    (df['Русский язык'] > russian_mean) &
    (df['АиНМА / Математика'] < math_mean) &
    (df['ИТ / Информатика и ИКТ / Физика'] < it_mean)
)

type1_count = condition_type1.sum()
type2_count = condition_type2.sum()

# Создаем итоговый DataFrame
summary_df = pd.DataFrame({
    "Подразделение": ["Факультет прикладной математики и механики"],
    "Уровень подготовки": ["Бакалавриат"],
    "Направление подготовки/специальность": ["Информационные системы и технологии"],
    "Год": [2025],
    "Количество мест": [62],
    "Предметы ЕГЭ": ["Математика, ИТ/Информатика/Физика, Русский язык"],
    "Средняя сумма баллов": [round(mean_total_score, 2)],
    "Min сумма баллов": [min_total_score],
    "Max сумма баллов": [max_total_score],
    "Предмет с высшим средним": [best_subject],
    "Кол-во студентов тип_1": [type1_count],
    "Кол-во студентов тип_2": [type2_count]
})

print("Итоговая таблица:")
display(summary_df)
```
